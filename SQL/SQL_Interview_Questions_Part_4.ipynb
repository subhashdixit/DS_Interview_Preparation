### **Ranking Functions**
* There are multiple functions that can be used for ranking the rows. When the value is the same, Rank() function will display the same rank value but the next value after that will be equal to the row number for example 1,1,3, and so on if the first two values are the same.
* Whereas in dense_rank() we values are continuous. For example, 1,1,2 and so on if the first two values are the same
* In the row_number() function the values are according to the row value 1,2,3 and so on even if the first two values are the same.

**Rank Function**

**Question Statement:**
* Calculate the rank according to the hire date of the employee and order by the salary of the employee**

```python
pd.read_sql('''
            SELECT 
            *, 
            RANK() OVER(ORDER BY HIRE_DATE) "RANK"
            FROM emp ORDER BY SALARY
            ''', db).head(10)
```

**Dense_Rank**

**Question Statement:**
* Calculate the rank according to the salary of the employee and order by the salary of the employee in ascending order

```python
pd.read_sql('''
            SELECT 
            *, 
            RANK() OVER(ORDER BY SALARY ) "RANK", 
            DENSE_RANK() OVER(ORDER BY SALARY) "DENSE_RANK",
            ROW_NUMBER() OVER(ORDER BY SALARY ) "ROW_NUMBER"
            FROM emp ORDER BY SALARY
            ''', db).head(10)
```

**Row_Number**
**Question Statement:**
* Calculate the rank according to the salary of the employee and order by the salary of the employee in ascending order. Use EmpId when sal is the same.

```python
pd.read_sql('''
            SELECT 
            *, 
            RANK() OVER(ORDER BY SALARY, EMPLOYEE_ID) "RANK", 
            DENSE_RANK() OVER(ORDER BY SALARY, EMPLOYEE_ID) "DENSE_RANK",
            ROW_NUMBER() OVER(ORDER BY SALARY,EMPLOYEE_ID) "ROW_NUMBER"
            FROM emp ORDER BY SALARY
            ''', db).head(10)
```

**Percent_Rank, Cume_Dist, NTILE**
**Problem Statement:**
* Show me distributions of Employee salary using PERCENT_RANK, CUME_DIST, and NTILE (quartile, decile, tertile, etc)

```python
pd.read_sql('''
            SELECT 
            *, 
            PERCENT_RANK() OVER(ORDER BY SALARY) "PERCENT_RANK",
            CUME_DIST() OVER(ORDER BY SALARY) "CUME_DIST",
            NTILE(4) OVER(ORDER BY SALARY) "NTILE"
            FROM emp ORDER BY SALARY
            ''', db).head(10)
```

### **Partition Clause**
* Partition clause is similar to group by however while both clauses aggregates data, Partition by would not reduce the number of rows while group by rolls up the data. partition by simply creates a window on which we can perform our calculations.

**Problem statement:**
* Find departmental salaries running total by employee name

```python
pd.read_sql('''
            SELECT 
            *, 
            SUM(SALARY) OVER(PARTITION BY DEPARTMENT_ID ORDER BY FIRST_NAME) as Running_Total
            FROM emp ORDER BY DEPARTMENT_ID, FIRST_NAME
            ''', db).head(10)
```

### **Window aggregation**
* We will start using windows clauses now. We can perform mathematical operations on the range of rows.

**Problem statement:**
* Find running total of total salary across the dept, largest to smallest

```python
pd.read_sql('''
            SELECT 
            *, 
            SUM(SALARY) OVER(ORDER BY DEPARTMENT_ID DESC ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) as Running_Total
            FROM emp ORDER BY DEPARTMENT_ID DESC
            ''', db).head(10)
```

**Problem statement:**
* Find running total of total salary for JOB_ID across the dept, largest to smallest

```python
pd.read_sql('''
            SELECT 
            *, 
            SUM(SALARY) OVER(PARTITION BY JOB_ID ORDER BY DEPARTMENT_ID DESC ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING) as Running_Total
            FROM emp ORDER BY DEPARTMENT_ID DESC
            ''', db).head(10)
```

**Problem statement:**
* 2 days moving average of salary per day.



```python
pd.read_sql('''
            SELECT 
            *, 
            AVG(SALARY) OVER(ORDER BY HIRE_DATE RANGE BETWEEN 2 PRECEDING AND CURRENT ROW) as Moving_Average
            FROM emp ORDER BY HIRE_DATE
            ''', db).head(10)
```

**Problem statement:** 
* Count the employees according to departments.

```python
# Simple way to find
pd.read_sql('''
            SELECT JOB_ID, 
            COUNT(JOB_ID)
            FROM
            emp
            GROUP BY JOB_ID
            ORDER BY JOB_ID
            ''', db).head(10)

```

```python
# Using last_value
pd.read_sql('''
            SELECT JOB_ID, 
            COUNT(DEPT)
            FROM(
            SELECT 
            *, 
            LAST_VALUE(JOB_ID) OVER(ORDER BY JOB_ID) as DEPT
            FROM emp ORDER BY HIRE_DATE
            )
            GROUP BY DEPT
            ORDER BY DEPT
            ''', db).head(10)

```

### **Lag and Lead**
* This is one of the most important windows functions.

**LAG**
* The lag function will just provide the data of the one row above.
* LAG(expression [,offset[,default_value]]) OVER(ORDER BY columns)

```python
pd.read_sql('''
            SELECT 
            *, 
            LAG(SALARY) OVER(ORDER BY SALARY) as PREVIOUS_SALARY
            FROM emp ORDER BY SALARY
            ''', db).head(10)
```

**LEAD**
* The lead function will do the opposite. It will provide the values of 1 row below.

```python
pd.read_sql('''
            SELECT 
            *, 
            LEAD(SALARY) OVER(ORDER BY SALARY) as PREVIOUS_SALARY
            FROM emp ORDER BY SALARY
            ''', db).tail(10)
```
